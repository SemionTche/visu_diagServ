```python
import zmq
```

```python
context = zmq.Context()
socket = context.socket(zmq.REQ)
socket.connect("tcp://localhost:1230")
```

```python
# send message to get answer
socket.send_string("__GET__")

reply = socket.recv_string()

print("Server answer :", reply)
```

```python
socket.send_string("__STOP__")
```

```python
socket.close()
```
